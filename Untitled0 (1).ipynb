```python
from sklearn.metrics import (confusion_matrix, classification_report, cohen_kappa_score,
                             mean_squared_error, mean_absolute_error, r2_score)

# Define a threshold to convert continuous 'ESTIMATE' into classes
threshold = np.median(y_train)

# Convert y_test into binary classes based on the threshold
y_test_class = (y_test > threshold).astype(int)

# Initialize a dictionary to store all metrics for each model
all_metrics = {}

# Loop through each model to compute the required metrics
for name, model in models.items():
    # Make predictions
    y_pred = model.predict(X_test)

    # Convert predictions to binary classes
    y_pred_class = (y_pred > threshold).astype(int)

    # Calculate confusion matrix
    conf_matrix = confusion_matrix(y_test_class, y_pred_class)
    correctly_classified = conf_matrix[0, 0] + conf_matrix[1, 1]
    incorrectly_classified = conf_matrix[0, 1] + conf_matrix[1, 0]

    # Calculate Kappa statistic
    kappa = cohen_kappa_score(y_test_class, y_pred_class)

    # Calculate error metrics
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    # Calculate relative errors
    mean_actual = np.mean(y_test)
    std_actual = np.std(y_test)
    relative_absolute_error = mae / mean_actual
    root_relative_squared_error = rmse / std_actual

    # Store all metrics in the dictionary
    all_metrics[name] = {
        "Correctly Classified Instances": correctly_classified,
        "Incorrectly Classified Instances": incorrectly_classified,
        "Kappa statistic": kappa,
        "Mean Absolute Error": mae,
        "Root Mean Squared Error": rmse,
        "Relative Absolute Error": relative_absolute_error,
        "Root Relative Squared Error": root_relative_squared_error,
        "Total Number of Instances": len(y_test)
    }

# Print all metrics for each model
for model_name, metrics in all_metrics.items():
    print(f"\n{model_name} Metrics:")
    for metric, value in metrics.items():
        print(f"{metric}: {value}")
```

```python
from sklearn.metrics import (precision_score, recall_score, f1_score, matthews_corrcoef,
                             roc_auc_score, precision_recall_curve, confusion_matrix)

# Loop through each model to compute classification metrics
for name, model in models.items():
    # Make predictions
    y_pred = model.predict(X_test)

    # Convert predictions to binary classes
    y_pred_class = (y_pred > threshold).astype(int)

    # Calculate confusion matrix components
    conf_matrix = confusion_matrix(y_test_class, y_pred_class)
    tn, fp, fn, tp = conf_matrix.ravel()  # Unpack the confusion matrix components

    # Calculate TP Rate (Recall) and FP Rate
    tp_rate = tp / (tp + fn) if (tp + fn) != 0 else 0
    fp_rate = fp / (fp + tn) if (fp + tn) != 0 else 0

    # Calculate other classification metrics
    precision = precision_score(y_test_class, y_pred_class, average='weighted')
    recall = recall_score(y_test_class, y_pred_class, average='weighted')  # Recall is the TP Rate
    f1 = f1_score(y_test_class, y_pred_class, average='weighted')
    mcc = matthews_corrcoef(y_test_class, y_pred_class)
    roc_area = roc_auc_score(y_test_class, y_pred_class)

    # Precision-Recall Curve Area
    precision_curve, recall_curve, _ = precision_recall_curve(y_test_class, y_pred_class)
    prc_area = np.trapz(recall_curve, precision_curve)

    # Print the Detailed Accuracy By Class
    print(f"\n{name} Detailed Accuracy By Class:")
    print(f"TP Rate (Weighted Avg.): {tp_rate}")
    print(f"FP Rate (Weighted Avg.): {fp_rate}")
    print(f"Precision (Weighted Avg.): {precision}")
    print(f"Recall (Weighted Avg.): {recall}")
    print(f"F-Measure (Weighted Avg.): {f1}")
    print(f"Matthews Correlation Coefficient (MCC): {mcc}")
    print(f"ROC Area (Weighted Avg.): {roc_area}")
    print(f"PRC Area (Weighted Avg.): {prc_area}")
```
